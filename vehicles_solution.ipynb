Task: Determine on which gear the car is going at the moment.
    The idea is to develop an algorithm that for a given observation will assign the gear the car was on at the moment.

Library import and data preparation:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
path = '/obd2.csv'
dane = pd.read_csv(path)
```

```python
dane.info()
```

We divide data into two sets: with rotations and with speeds

```python
dane1 = dane.loc[dane['type'] == 'rpm']
dane2 = dane.loc[dane['type'] == 'speed']
```

We need rotations and speed for the same vehicle and time, so we combine data 1 and data 2
with the vehicle_id and timestamp key

```python
result = pd.merge(dane1, dane2, on = ['vehicle_id','ride_id','timestamp'])
```

We failed to connect everyone, but if we do not have both: rotations and speed, we can't do much, so we reject such observations for now.
Renaming columns:

```python
result.columns.values[4] = 'rpm'
result.columns.values[6] = 'speed'
del result['type_x']
del result['type_y']
```

For now, we are throwing out some disturbing records that will have to be handled later. Following values are considered “disturbing”:
- Rotations below min_rpm - we assume that a car cannot go at such rotations, so it is an idle.
- Speed >= max_speed - outliers that can make it difficult to find a model, later we will assign the highest gear to them, provided that the number of turns is reasonable
- Speed <= min_speed, as above, later we will assign the lowest gear to them, provided that
the number of turns is reasonable

```python
min_rpm = 900
min_speed = 10
max_speed = 200
result0 = result.loc[result['rpm'] <= min_rpm]
result = result.loc[result['rpm'] > min_rpm]
result_min_speed = result.loc[result['speed'] <= min_speed]
result = result.loc[result['speed'] > min_speed]
result_max_speed = result.loc[result['speed'] >= max_speed]
result = result.loc[result['speed'] < max_speed]
```

The relation between RPM and speed is a linear relationship constant for a given car and for a
given gear (wheel circumference and gear ratio). We are interested in finding clusters
composed of points (rpm, speed) lying on straight lines passing through the point (0,0).
Because for each car there may be other transmissions and different circumferences of
wheels, so we must treat each car separately. Let's see what the situation looks like for the
first car.

```python
result1 = result.loc[result['vehicle_id'] == 1] 
```

```python
plt.scatter(result1.speed,result1.rpm)
plt.show()
```

As we can see on the chart, the car has 6 gears.

 We reduce the two-dimensional problem to a 1-dimensional one:

```python
result['iloraz'] = result.rpm/result.speed
```

The first attempt to find clusters (KMeans) showed that the clusters centers corresponding to the directional coefficients of individual straight lines are growing exponentially. The use of double logarithms causes that the distances between the centers of classes corresponding to subsequent gears become similar.

```python
result['potega'] = np.log(np.log(result.iloraz))
```

```python
x = range(1,7)
a = [18.07,  112.47,   39.37,   61.83,  27.43,   21.61] #cluster centers found before
a.sort()
c = [np.log(np.log(y)) for y in a ]
plt.subplot(1, 2, 1)
plt.scatter(x,a)
plt.title('Raw cluster centers')
plt.ylabel('rpm/speed')
plt.xlabel('cluster number')
plt.subplot(1, 2, 2)
plt.scatter(x,c)
plt.title('Log log cluster centers')
plt.ylabel('log(log(rpm/speed))')
plt.xlabel('cluster number')
plt.tight_layout(pad=1, w_pad=10, h_pad=10.0)
plt.show()  
```

```python
n_cars = result.vehicle_id.value_counts().shape[0]
```

Procedure:
We're taking cars one by one in a loop.
First we need to determine if there are 5 or 6 gears for the given car (we assume that these are the only possible values).
Because for higher rotations the points are more closely situated on straight lines, we will check the number of gears on the part of the set, where the rotations number is higher than the median number for the given car.
For this set, we fit models for the number of clusters from 4 to 7 and remember the inertia for each of them.
Next, we check whether the smaller refraction angle is for 5 or 6 clusters and thus establish the number of clusters (gears).
The next step is to determine the initial clusters centers, also based on the set with higher rotations (in case of random initial centers, model may find wrong centers).
In the end, we fit the final model for the whole set (without records deleted first) initiating the search from the centers found previously.
Assigned labels are attached as a column to the data frame, and then concatenate.

```python
def pad(length, list_):
    """
    Appends zeros to list to get length of list equal to `length`.
    """
    if len(list_) < length:
        zeros = length - len(list_)
        return (list_ + ['NaN'] * zeros )
    else:
        return list_
```

```python
m=3
ks = range(m,8)
col = np.array(["blue","green",'yellow', "orange","red","purple"])
#creating empty data frame
con_result = result.loc[result['vehicle_id'] == 100]
con_result['labels'] = []
#creating empty data frame for centers
columns = [1,2,3,4,5,6]
all_centers = pd.DataFrame(columns=columns)
#creating list to remember number of gears for each vehicle
car_gears = []
# vehicle loop 
for i in range(1,n_cars+1):
    result1 = result.loc[result['vehicle_id'] == i]
    iloraz = np.array([result1.potega])
    iloraz=iloraz.T
    # we remember the inertia for different gear numbers
    inertia = []
    for k in ks:
        model = KMeans(n_clusters=k)
        model.fit(iloraz)
        inertia.append(model.inertia_)
    print("vehicle_id=",i)
    plt.plot(ks,inertia) 
    plt.show() 
    # we are looking for the "elbow", where the graph breaks at the slightest angle
    val = []
    for j in range(0,(7-m-1)):
        a = (inertia[j]-inertia[j+1])*10
        b = (inertia[j+1]-inertia[j+2])*10
        val.append ((a-b)/ np.sqrt((a**2+1)*(b**2+1)))
    d = np.argmax(val)
    gears = d+4
    car_gears.append([i,gears])
    print('val',val)
    print("number of gears", gears)
    
    model = KMeans(n_clusters=gears)
    model.fit(iloraz)
    labels = model.predict(iloraz)
    centers = model.cluster_centers_
    cent = []
    for y in centers:
        cent.append(y[0])
    cent.sort()
    cent = pad(6,cent)
    cent = pd.DataFrame([cent],columns=columns)
    all_centers = pd.concat([all_centers,cent])
    plt.scatter(result1.rpm,result1.speed,c = col[labels])
    plt.show()  
    result1['labels'] = gears - labels
    con_result= pd.concat([con_result,result1])
  
```

It can be seen that the ratios rpm/speed are different for individual cars

```python
result_max_speed = pd.merge(result_max_speed, car_gears,on = ['vehicle_id'])
```

The final result with attached disturbing records:

```python
con_result= pd.concat([con_result,result0])
con_result= pd.concat([con_result,result_min_speed])
con_result= pd.concat([con_result,result_max_speed])
```
